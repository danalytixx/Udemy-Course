```python
from google.colab import userdata
```

```python
import requests
import pandas as pd

import re

# Define the URL and headers
url = userdata.get('URL')
headers = {
    'Accept': 'application/json',
    'AuthenticationToken': userdata.get('ATH_TOKEN'),
    'Cookie':  userdata.get('COOKIE')
}

# Function to clean text
def clean_text(text):
    text = re.sub(r'\\u[0-9a-fA-F]{4}', '', text)  # Remove \u unicode characters
    text = text.replace('\n', '')  # Remove newline characters
    text = text.replace('’', '')  # Remove special characters
    text = text.replace('„', '')  # Remove special characters
    text = text.replace('“', '')  # Remove special characters
    return text

# Make the GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    response_json = response.json()

    # Extract the required fields and clean the text
    name = clean_text(response_json.get('name', ''))

    # Extract specific attributes
    attributes = response_json.get('customAttributes', [])
    attribute_166589 = ''
    attribute_149011 = ''
    attribute_149014 = ''
    attribute_149017 = ''
    attribute_149020 = ''
    attribute_148990 = ''
    attribute_2513 = ''
    attribute_166613 = ''
    attribute_145955 = ''
    attribute_145958 = ''
    attribute_145961 = ''
    attribute_145964 = ''
    attribute_145967 = ''
    attribute_148988 = ''

    for attribute in attributes:
        if attribute['attributeDefinitionId'] == '166589':
            attribute_166589 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '149011':
            attribute_149011 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '149014':
            attribute_149014 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '149017':
            attribute_149017 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '149020':
            attribute_149020 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '148990':
            attribute_148990 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '2513':
            attribute_2513 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '166613':
            attribute_166613 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '145955':
            attribute_145955 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '145958':
            attribute_145958 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '145961':
            attribute_145961 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '145964':
            attribute_145964 = clean_text(attribute.get('stringValue', ''))
        elif attribute['attributeDefinitionId'] == '145967':
            attribute_145967 = clean_text(str(attribute.get('numberValue', '')))
        elif attribute['attributeDefinitionId'] == '148988':
            attribute_148988 = clean_text(attribute.get('stringValue', ''))

    # Create a DataFrame
    data = {
        'Name': name,
        'Attribute 166589': attribute_166589,
        'Attribute 149011': attribute_149011,
        'Attribute 149014': attribute_149014,
        'Attribute 149017': attribute_149017,
        'Attribute 149020': attribute_149020,
        'Attribute 148990': attribute_148990,
        'Attribute 2513': attribute_2513,
        'Attribute 166613': attribute_166613,
        'Attribute 145955': attribute_145955,
        'Attribute 145958': attribute_145958,
        'Attribute 145961': attribute_145961,
        'Attribute 145964': attribute_145964,
        'Attribute 145967': attribute_145967,
        'Attribute 148988': attribute_148988
    }
    df = pd.DataFrame([data])


else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
    print(f"Response: {response.text}")
```
