```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
df = tesla.history(period="5y")

df.head()
```

```python
# Data Cleaning
print(df.isnull().sum())
print(df.dtypes)
```

```python


# Data Exploration
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)
```

```python


# Plot the historical closing prices
plt.figure(figsize=(12, 6))
plt.plot(df['Close'])
plt.title('Tesla Closing Price History')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()

# Feature Engineering
df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()
df.head()

```

```python

# Visualization

# Closing Price with Moving Averages
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close')
plt.plot(df['MA50'], label='50-Day MA')
plt.plot(df['MA200'], label='200-Day MA')
plt.title('Tesla Closing Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Volume Traded
plt.figure(figsize=(12, 6))
plt.plot(df['Volume'])
plt.title('Tesla Volume Traded')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()

# Monthly Returns
df['Monthly Return'] = df['Close'].resample('M').ffill().pct_change()

plt.figure(figsize=(12, 6))
plt.plot(df['Monthly Return'])
plt.title('Tesla Monthly Returns')
plt.xlabel('Date')
plt.ylabel('Monthly Return')
plt.show()

# Distribution of Daily Returns
df['Daily Return'] = df['Close'].pct_change()

plt.figure(figsize=(12, 6))
sns.histplot(df['Daily Return'].dropna(), bins=50, kde=True)
plt.title('Distribution of Tesla Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python

# Box Plot of Monthly Returns
monthly_returns = df['Monthly Return'].dropna()
monthly_returns = monthly_returns.reset_index()
monthly_returns['Month'] = monthly_returns['Date'].dt.to_period('M')
monthly_returns['Year'] = monthly_returns['Date'].dt.to_period('Y')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='Monthly Return', data=monthly_returns)
plt.xticks(rotation=90)
plt.title('Box Plot of Tesla Monthly Returns')
plt.xlabel('Month')
plt.ylabel('Monthly Return')
plt.show()
```
