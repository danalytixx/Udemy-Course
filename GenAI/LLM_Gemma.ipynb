```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Set your Hugging Face token
hf_token = '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'

# Load tokenizer and model using the token
tokenizer = AutoTokenizer.from_pretrained("google/gemma-7b", use_auth_token=hf_token)
model = AutoModelForCausalLM.from_pretrained("google/gemma-7b", use_auth_token=hf_token)

# Input text
input_text = "Write me a poem about Machine Learning."
input_ids = tokenizer(input_text, return_tensors="pt").input_ids

# Generate output with a specified maximum length
outputs = model.generate(input_ids, max_new_tokens=50)

# Decode and print the result
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
