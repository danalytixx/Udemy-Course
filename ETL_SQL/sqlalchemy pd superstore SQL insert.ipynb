```python
import pandas as pd
from sqlalchemy import create_engine, Table, Column, Integer, String, Float, DateTime, MetaData
from sqlalchemy.orm import sessionmaker

# Load the CSV file into a DataFrame
file_path = 'C:/Users/User/Desktop/Udemy Course/Sample - Superstore.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows to understand the structure
print(data.head(10))

# Define the connection parameters
server = 'HM-PC\\SQLEXPRESS2024'
database = 'DATA'
username = 'admin'
password = 'admin2024'

# Create a connection string
connection_string = f'mssql+pyodbc://{username}:{password}@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server'

# Create a SQLAlchemy engine
engine = create_engine(connection_string)

# Define the metadata
metadata = MetaData()

# Define the table structure
superstore_table = Table('Superstore', metadata,
    Column('Order ID', String(255)),
    Column('Order Date', DateTime),
    Column('Ship Date', DateTime),
    Column('Ship Mode', String(255)),
    Column('Customer ID', String(255)),
    Column('Customer Name', String(255)),
    Column('Segment', String(255)),
    Column('Country', String(255)),
    Column('City', String(255)),
    Column('State', String(255)),
    Column('Postal Code', String(255)),
    Column('Region', String(255)),
    Column('Product ID', String(255)),
    Column('Category', String(255)),
    Column('Sub-Category', String(255)),
    Column('Product Name', String(255)),
    Column('Sales', Float),
    Column('Quantity', Integer),
    Column('Discount', Float),
    Column('Profit', Float)
)

# Drop the existing table if it exists
metadata.drop_all(engine, [superstore_table])

# Create the table in the database
metadata.create_all(engine)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()

# Insert data in chunks
chunk_size = 10000
for start in range(0, len(data), chunk_size):
    chunk = data.iloc[start:start + chunk_size]
    records = chunk.to_dict(orient='records')
    session.execute(superstore_table.insert(), records)
    session.commit()

# Close the session
session.close()

print("Data has been successfully written to the SQL Server database.")
```
